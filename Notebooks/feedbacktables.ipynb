```python
def to_table(idf, columns):
    cols = [0, *columns]
    name = f"Antworten auf die Frage: {idf.columns[columns[0]]}"
    df = idf.iloc[:, cols].copy()
    if len(columns) == 2:
        df.columns = ["Person", "Antwort", "Anmerkung"]
    else:
        df.columns = ["Person", "Antwort"]

    return f"""
\\begin{{table}}[ht]
\centering
\\resizebox{{\\textwidth}}{{!}}{{
{df.to_latex(index=False)}
}}
    \caption{{{name}}}
    \label{{tb:{columns[0]}}}
\end{{table}}
    """
```

```python
for i in range(2,11):
    print(to_table(df, [i]))
```

```python
df.fillna('',inplace=True)
for i in range(11, df.shape[1], 2):
    # print(df.columns[i], df.columns[i+1])
    print(to_table(df, [i, i+1]))
```
